```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/student-mat.csv')
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
columns_to_encode = df.select_dtypes(include=['object']).columns.tolist()
```

```python
df[columns_to_encode] = df[columns_to_encode].apply(lambda col: label_encoder.fit_transform(col))
```

```python
df.corr()
```

```python
df.shape
```

```python
X = df.iloc[:,:32]
y = df["G3"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train,y_train)
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pre = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pre)
```

```python
from sklearn.linear_model import Lasso
ls = Lasso(alpha=0.2)  # Adjust alpha for regularization strength
ls.fit(X_train, y_train)
y_pred = ls.predict(X_test)
r2_score(y_test, y_pred)
```

```python
coefs = []
r2_scores = []
for i in [0,0.1,0.2,10,20]:
  rg = Lasso(alpha=i)
  rg.fit(X_train, y_train)
  coefs.append(rg.coef_.tolist())
  y_pred = rg.predict(X_test)
  r2_scores.append(r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(12,6))
plt.subplot(221)
plt.bar(X.columns, coefs[0])
plt.xticks(rotation=90)
plt.title("alpha = 0,r2_score={}".format(round(r2_scores[0],2)))
plt.show()

plt.figure(figsize=(12,6))
plt.subplot(222)
plt.bar(X.columns, coefs[1])
plt.xticks(rotation=90)
plt.title("alpha = 0.1,r2_score={}".format(round(r2_scores[1],2)))
plt.show()

plt.figure(figsize=(12,6))
plt.subplot(223)
plt.bar(X.columns, coefs[2])
plt.xticks(rotation=90)
plt.title("alpha = 0.2,r2_score={}".format(round(r2_scores[2],2)))
plt.show()

plt.figure(figsize=(12,6))
plt.subplot(224)
plt.bar(X.columns, coefs[3])
plt.xticks(rotation=90)
plt.title("alpha = 10,r2_score={}".format(round(r2_scores[3],2)))
plt.show()
```

```python
alphas = [0,0.0001,0.001,0.01,0.1,0.2,1,10,100]
coefs = []
for i in alphas:
  rg = Lasso(alpha=i)
  rg.fit(X_train, y_train)
  coefs.append(rg.coef_.tolist())
```

```python
array = np.array(coefs)
dc = pd.DataFrame(array,columns = X.columns)
dc["alpha"] = alphas
dc.set_index("alpha")
```
