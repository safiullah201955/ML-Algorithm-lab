```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/CarPrice_Assignment.csv")
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
columns_to_encode = df.select_dtypes(include=['object']).columns.tolist()
```

```python
df[columns_to_encode] = df[columns_to_encode].apply(lambda col: label_encoder.fit_transform(col))
```

```python
df.head(3)
```

```python
df.corr()
```

```python
X = df.iloc[:,:25]
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
Lg = LinearRegression()
Lg.fit(X_train,y_train)
y_pred = Lg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
reg = Ridge(alpha=0.05)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
reg1 = Lasso(alpha=0.01)
reg1.fit(X_train,y_train)
y_pred = reg1.predict(X_test)
r2_score(y_test,y_pred)
```

```python
reg2 = ElasticNet(alpha=1.7,l1_ratio=0.9)
reg2.fit(X_train,y_train)
y_pred = reg2.predict(X_test)
r2_score(y_test,y_pred)
```
